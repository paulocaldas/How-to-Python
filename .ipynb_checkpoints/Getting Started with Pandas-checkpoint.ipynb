```python
# import the basics

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

**READ DATA AS A TABLE**

```python
filename = 'data_examples/FtsZ Alone 2018.01.16 ExpI_growth_tracks.csv' # path to file with extension!

mytable = pd.read_csv(filename)     # reads the data as a table
mytable.head(10)                    # check how the first 20 rows look like
```

**PANDAS BASICS**

```python
# Select a specific column
track_dur = mytable['TRACK_DURATION']

#select a line number
track27 = mytable.iloc[27]

# Select a subset of the sable (create a new table)
new_table = mytable[['Label','TRACK_DURATION','TRACK_MEDIAN_SPEED']]

# change column names
new_table.columns = ['Track', 'Duration', 'Speed']

#check the new table
new_table.head(10) #check the head of the new table
```

**PLOT USING PANDAS COLUMNS**

```python
# BASIC HISTOGRAM PLOTTING
speed = new_table['Speed']  # data to plot
plt.hist(speed) ;           # basic plotting
```

```python
# BASIC XY PLOT

x = mytable['NUMBER_SPOTS']
y = mytable['TRACK_DURATION']
plt.scatter(x,y);
plt.plot(x,y)
```

**IMPROVE YOUR HISTOGRAM** <br>
a few command examples <br>
much more options to explore .... google!

```python
plt.figure(figsize=(4,3), dpi =150)  # set figure size and dpi to make it bigger (optional)

plt.hist(speed, bins = 20, color = 'seagreen', edgecolor = 'black', label = 'FtsZ')
plt.legend()

plt.xlim([0,0.2])
plt.ylim([0,3000])
plt.xlabel('Velocity (um/s)', fontsize=11)
plt.ylabel('Counts', fontsize=11)
plt.title('Trajectory Speed')
plt.tick_params(direction = 'in', top=False, right = False) 

# SAVE THE FIGURE
plt.savefig('figure_name.png', format = 'png', dpi = 150, bbbox_inches = 'tight') # save figure same directory
```

**IMPROVE YOUR XY PLOT** <br>

```python
x = mytable['NUMBER_SPOTS']
y = mytable['TRACK_DURATION']

x = x[ :50] #clipping the data
y = y[ :50]

plt.figure(figsize=(4,3), dpi =150)

plt.plot(x, y, '-o', color = 'black', linewidth = 0.5, 
         markersize = 5, markeredgecolor = 'black', markerfacecolor = 'lightblue')

# a few parameter to format
plt.xlabel('number of spots per track', fontsize = 12)
plt.ylabel('Duration (s)', fontsize=12)
plt.title('dummy plot', fontsize = 12)

#plt.legend(loc=2, frameon=False, fontsize = 10)
plt.grid(False)
plt.tick_params(top=False, right=False)
plt.tick_params(direction = 'in', top=False, right=False)
plt.grid(color='gray', alpha=0.2, axis = 'y', ls = 'dashed')

#plt.xlim([0,25])
plt.ylim([0,40]);
```

**BASIC BOXPLOT** <br>
a bit more tricky already

```python
plt.figure(figsize=(4,3), dpi =150)

mean_quality = mytable['TRACK_MEAN_QUALITY']
max_quality = mytable['TRACK_MAX_QUALITY']
min_quality = mytable['TRACK_MIN_QUALITY']

myboxplot = plt.boxplot([mean_quality, min_quality, max_quality], patch_artist = True, 
                        notch = True, widths = 0.3, sym = 'o',
                        labels = ['mean','max','min'])

plt.xlabel('Quality Parameters', fontsize=9)
plt.ylabel('Number of Tracks', fontsize=11)
#plt.title('')

#set the color
colors = ['blue', 'seagreen', 'orange']

# use a for loop to define colors and other parameters for each boxpot ...

for patch, color in zip(myboxplot['boxes'], colors):
    patch.set_color(color)
    patch.set_edgecolor('black')
    patch.set_linewidth(0.5)

# set other global settings like:

plt.setp(myboxplot['fliers'], marker='o', color='black', alpha=0.5, markersize= 3)
plt.setp(myboxplot['medians'], color = 'white');
    
# format a boxplot for the first time takes time, you should google for much more parameters according to your needs ...
```

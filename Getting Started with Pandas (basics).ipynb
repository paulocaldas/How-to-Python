```python
# import the basics

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

**Read data with pandas**

```python
filename = 'data_examples/FtsZ Alone 2018.01.16 ExpI_growth_tracks.csv' # path to file with extension!

mytable = pd.read_csv(filename)     # reads the data as a table (data type is a pandas dataframe)
mytable.head(10)                    # check how the first 20 rows look like
```

**Examples of Data Selection and Operations with Pandas**

```python
track_duration = mytable['TRACK_DURATION']  # select a column as a list (panda series)
track27 = mytable.iloc[27]                  # select a line

# quick math with pandas
track_duration.dropna()                 # drop all NA values from the series 
track_duration.mean()                   # get the mean of the column
track_duration.std()                    # get the stdev of the column
track_duration.max()
track_duration.min()

# select chucks of data
track_duration[:10]                     # select first 10 lines
track_duration[40:50]                   # select a slice of the rows
```

```python
# take the mean of all lines at once
mytable.mean(1)

# take the mean of all columns at once
mytable.mean()
```

**Data Filtering**

```python
mytable[track_duration > 35]       # filter all rows with track duration lower than 35 sec
```

```python
mytable[track_duration > 35]       # filter all rows with track duration lower than 35 sec
```

**Create subtables**

```python
subtable = mytable[['TRACK_DURATION','TRACK_MEAN_SPEED']]  # new table with two columns
subtable.columns = ['Duration','Speed']                    # rename the columns 
subtable.head(10)                                          # check the header
```

**Add and Drop Columns**

```python
no_speed = subtable.drop('Speed', axis = 1)  # drop column named speed
no_speed.head(10)
```

```python
random_list = np.arange(0,len(subtable))  # creates list from 0 to table size

no_speed['NEW COL'] = random_list         # adds a new column
no_speed.head(15)
```

**Combine Different Files into a Table**

```python
filename1 = 'data_examples/ZapA_growth_tracks.xlsx'
filename2 = 'data_examples/ZapA_shrinkage_tracks.xlsx'

# read files from a specific sheet inside an excel book

speed_g = pd.read_excel(filename1, sheetname = 1, header = None)
speed_s = pd.read_excel(filename2, sheetname = 1, header = None)

# combine dataset by column
speed_table = pd.concat([speed_g,speed_s], axis = 1)
speed_table.columns = ['speed_g','speed_s']
speed_table.head(15)                                      
```

**Basic plotting using Pandas columns**

```python
# get the data by selecting columns or lines

speed = subtable['Speed'] 
duration = subtable['Duration']

# plot an histogram
plt.figure(figsize=(4,3), dpi =150)
plt.hist(duration, bins = 20, color = 'orange', edgecolor = 'black', label = 'duration')
plt.legend()
plt.xlabel('duration (s)')
plt.ylabel('counts')

#plot a second histogram in a different window
plt.figure(figsize=(4,3), dpi =150)
plt.hist(speed, bins = 20, color = 'seagreen', edgecolor = 'black', label = 'duration')
plt.legend()
plt.xlabel('speed (um/s)')
plt.ylabel('counts')
```

```python
x = mytable['NUMBER_SPOTS']
y = mytable['TRACK_DURATION']

x = x[ :50] # clipping the data
y = y[ :50]

plt.figure(figsize=(4,3), dpi =150)

plt.plot(x, y, '--o', color = 'black', linewidth = 0.2, 
         markersize = 6, markeredgecolor = 'black', markerfacecolor = 'lightblue')

# a few parameter to format
plt.xlabel('number of spots per track', fontsize = 12)
plt.ylabel('Duration (s)', fontsize=12)
plt.title('dummy plot', fontsize = 12)

#plt.legend(loc=2, frameon=False, fontsize = 10)
#plt.grid(False)
#plt.tick_params(top=False, right=False)
#plt.tick_params(direction = 'in', top=False, right=False)
#plt.grid(color='gray', alpha=0.2, axis = 'y', ls = 'dashed')

#plt.xlim([0,25])
#plt.ylim([0,40]);
```

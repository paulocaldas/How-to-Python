```python
import numpy as np
import pandas as pd
import scipy.stats
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style(style='white')
%matplotlib inline
```

**Plot Multiple Box Plots from a Data Frame**

```python
box_data = pd.read_table('data_examples/boxplots_data.txt')

plt.figure(figsize = (4,3), dpi = 150)
sns.boxplot(data = box_data,  whis = 2,  width = 0.6, palette = "colorblind" , linewidth = 1.5)
sns.stripplot(data = box_data, jitter=True, marker = 'o', size = 7,  alpha=0.5, color = 'black')

plt.tick_params(top=False, right=False)
plt.ylabel('Velocity (nm/s)', fontsize = 12)
plt.tick_params(direction = 'in', top=False, right = False)
#plt.xticks(np.arange(3),['Kymographs','Track Displacement','Differential Image'], fontsize = 10, rotation=30);
#plt.ylim([30,110])

plt.savefig('plot_examples/box_plot_example.pdf', transparent=False, bbox_inches = 'tight')
```

**Two-sided t-test**

```python
#statistical two-sided t-test for two samples

FtsZ_sample = box_data['FtsZ'].dropna()
ZapA_sample = box_data['ZapA wt'].dropna()

t_statistic, pval = scipy.stats.ttest_ind(FtsZ_sample,ZapA_sample, equal_var=False)
'pval = {:4.2}'.format(pval)
```
